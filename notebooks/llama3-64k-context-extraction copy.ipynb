```python
import requests

url = "https://api.firecrawl.dev/v0/search"
query = "Nvidia stock outlook"
payload = {
    "query": query,
    "pageOptions": {
        "onlyMainContent": True,
        "fetchPageContent": True
    },
    "searchOptions": {"limit": 4}
}
headers = {
    "Authorization": "",
    "Content-Type": "application/json"
}

response = requests.request("POST", url, json=payload, headers=headers)

print(response.text)
```

```python
import json

# Parse the JSON response from the API
response_data = json.loads(response.text)

# Check if the response was successful
if response_data.get("success", False):
    # Initialize an empty string to hold all the concatenated markdown content
    concatenated_markdown = ""
    
    # Iterate through each item in the data list
    for item in response_data.get("data", []):
        # Concatenate the markdown content of each item
        concatenated_markdown += item.get("markdown", "") + "\n\n"
    
    # Print the concatenated markdown content
    print(concatenated_markdown)
else:
    print("Failed to retrieve data")


print(len(concatenated_markdown))
```

```python
# Count the number of tokens in concatenated_markdown
# We use the split method to count the words as a proxy for tokens
token_count = len(concatenated_markdown.split())
print("Number of words in concatenated_markdown:", token_count)
```
